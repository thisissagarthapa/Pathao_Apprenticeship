```python
np.random.seed(42)  # For reproducibility
dataset['new_outlier_column'] = np.random.normal(loc=50, scale=10, size=3353)  # Normal distribution

# Adding outliers: Inserting some extreme values (e.g., 100, 120) in random places
outlier_indices = np.random.choice(dataset.index, size=10, replace=False)  # Randomly select 10 rows
dataset.loc[outlier_indices, 'new_outlier_column'] = np.random.choice([100, 120, 150], size=10)

# Check the first few rows of the updated DataFrame
print(dataset.head())
```

```python
dataset.describe()
```

```python
sns.boxplot(x="new_outlier_column",data=dataset)
plt.show()
```

```python
min_range = dataset["new_outlier_column"].mean() - (3 * dataset["new_outlier_column"].std())
max_range = dataset["new_outlier_column"].mean() +(3 * dataset["new_outlier_column"].std())
```

```python
min_range,max_range
```

```python
new_data=dataset[dataset["new_outlier_column"]<=max_range]
```

```python
sns.boxplot(x="new_outlier_column",data=dataset)
plt.show()
```

```python
sns.boxplot(x="new_outlier_column",data=new_data)
plt.show()
```

```python
updated_dataset=new_data[new_data["new_outlier_column"]>=min_range]
```

```python

sns.boxplot(x="new_outlier_column",data=updated_dataset)
```

```python
updated_dataset
```

Z-Score


```python
z_score=((updated_dataset['new_outlier_column'])-(updated_dataset['new_outlier_column'].mean()))/(updated_dataset['new_outlier_column'].std())
```

```python
z_score
```

```python
updated_dataset['z_score']=z_score
```

```python
updated_dataset
```

```python
updated_dataset["z_score"]<3
```

```python
sns.boxplot(x="new_outlier_column",data=updated_dataset)
plt.show()
```
